```python
import pandas as pd
```

```python
df = pd.read_excel("/project/hackathon/hackers06/shared/pat_hsp.xlsx")
```

```python
def calculateLOS(df, admissionTimeCol, dischargeTimeCol):
    df[admissionTimeCol] = df[admissionTimeCol].astype('datetime64[ns]')
    df[dischargeTimeCol] = df[dischargeTimeCol].astype('datetime64[ns]')
    df["LOS"] = df.apply(lambda r: (r[dischargeTimeCol] - r[admissionTimeCol]).days, axis=1)
```

```python
calculateLOS(df, "HOSP_ADMSN_TIME", "HOSP_DISCH_TIME")
```
