```python
import pandas as pd
import pprint
```

```python
def getGroupMapping(filepath, irfCol, irfFilter, impairCol, impairDescCol, etiologicCol, etiologicDescCol):
    df = pd.read_excel(filepath)
    qualifyingAdmissions = df[df[irfCol] == irfFilter]
    impairmentGroupToIcdMap = {}
    impairmentNames = {}
    icd10 = {}
    for index, row in qualifyingAdmissions.iterrows():
        impairCode = row[impairCol]
        icdCode = row[etiologicCol]
        impairmentNames[impairCode] = row[impairDescCol]
        icd10[icdCode] = row[etiologicDescCol]
        if impairCode not in impairmentGroupToIcdMap:
            impairmentGroupToIcdMap[impairCode] = set()
        impairmentGroupToIcdMap[impairCode].add(icdCode)
        result = {}
    for code in impairmentGroupToIcdMap:
        if code not in result:
            result[(code, impairmentNames[code])] = {
                "icd10": []
            }
        for icd in impairmentGroupToIcdMap[code]:
            result[(code, impairmentNames[code])]["icd10"].append({
                "name": icd10[icd],
                "code": icd
            })
    return result
```

```python
getGroupMapping("uds_qualified_data.xlsx", "IRF", "Qualifying Admission impairment", "ADM_IMPAIR_CODE", "ADM_IMPAIR_DESC", "ETIOLOGIC_DX", "ETIOLOGIC_DX_DESC")
```
