```python
import pandas as pd
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from matplotlib.pyplot import figure
from collections import Counter
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE 
from imblearn.over_sampling import ADASYN 
from imblearn.over_sampling import SMOTENC
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SVMSMOTE
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import GridSearchCV
from imblearn.ensemble import BalancedBaggingClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
from imblearn.ensemble import RUSBoostClassifier
from sklearn.model_selection import GroupShuffleSplit

from imblearn.over_sampling import SMOTE 
from imblearn.over_sampling import ADASYN
from imblearn.over_sampling import KMeansSMOTE
from imblearn.over_sampling import BorderlineSMOTE
from imblearn.over_sampling import SMOTENC
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SVMSMOTE
from sklearn.metrics import precision_recall_fscore_support
```

```python
df = pd.read_csv("final.csv") 
```

```python
X_train, X_test, y_train, y_test = train_test_split( df.drop(['y_actual','IHAR','DISCH_DELEY'],axis=1)
   ,
    df['y_actual'],
    test_size=0.2,
    random_state=0)

X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=50)
rfc.fit(X_train,y_train)
```

```python
predictions = rfc.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
```

```python
plt.figure(figsize = (9.7,6))
sns.set(font_scale=2)
confusion_matrix = pd.crosstab(y_test, predictions, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True, annot_kws={"size": 18}, cmap='Blues', fmt='g')
```

```python
feature_importances = pd.DataFrame(rfc.feature_importances_,
                                  index = X_train.columns,
                                   columns=['importance']).sort_values('importance', ascending=False).reset_index()

sns.set(rc={'figure.figsize':(13.7,40)})
fig, ax = plt.subplots(dpi = 300)
sns.set_style("white")
sns.barplot(data=feature_importances.head(20), y='index', x='importance', palette='GnBu_d' )
```

## SMOTE : Approach

```python
smt = SMOTE()
ada = ADASYN()
ros= RandomOverSampler()
svs=SVMSMOTE()
bls=BorderlineSMOTE()

over_sampling={'SMOTE':smt,'ADASYN':ada,
               'RandomOverSampler':ros,'SVMSMOTE':svs,'BorderlineSMOTE':bls}
```

```python
rfc = RandomForestClassifier(n_jobs=2,max_features= 'sqrt' ,n_estimators=50, oob_score = True) 
param_grid = { 
    'n_estimators': [10, 20, 50, 80, 100, 200], 'max_features': ['auto', 'sqrt', 'log2']
   
}

Over_Sampling_Grid={}
for technique in over_sampling.keys():
    thistechnique=over_sampling[technique]
    print("Applying : {} : Over-Sampling Method \n".format(technique))
    X_train_res, y_train_res = thistechnique.fit_sample(X_train, y_train)
    CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 3,verbose=False)
    CV_rfc.fit(X_train_res,y_train_res)
    print("Best Params {} with {} over-sampling technique \n".format(CV_rfc.best_params_,technique))
    grid_predictions = CV_rfc.predict(X_test)
    Over_Sampling_Grid[technique]=grid_predictions
    
    
    
    
 
```

```python
for technique in Over_Sampling_Grid.keys():
    thisgridprediction=Over_Sampling_Grid[technique]
    print("Confusion Matrix for RF Classifier with {} over-sampling technique \n".format(technique))
    print(classification_report(y_test,thisgridprediction))
    plt.figure(figsize = (9.7,6))
    sns.set(font_scale=2)
    confusion_matrix = pd.crosstab(y_test, thisgridprediction, rownames=['Actual'], colnames=['Predicted'])
    sns.heatmap(confusion_matrix, annot=True, annot_kws={"size": 18}, cmap='Blues', fmt='g')
    
    plt.show()   
```
